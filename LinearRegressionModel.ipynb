```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data={'Average_daily_Usage_hrs': [3, 4, 6, 7, 8, 10, 12],
      'Monthly_Bill$': [50, 60, 80, 95,110, 130, 150]}
```

```python
df=pd.DataFrame(data)
df
```

```python
#Manually define Average_daily_Usage_hrs as X and Monthly_Bill$ as y
X=np.array([[3], [4], [6], [7], [8], [10], [12]])
y=np.array([50, 60, 80, 95, 110, 130, 150])
```

```python
#Initialize and train model
model= LinearRegression() 
model.fit(X,y)
```

```python
y_pred= model.predict(X)
y_pred
```

```python
#predicting monthly bill for a household that uses electricity for 5 hrs a day
new_input= 5
pred_bill_5hrs=model.predict([[5]])
print("predicted monthly bill for 5 hrs of daily usage:", pred_bill_5hrs[0])
```

```python
#The slope(rate of increase per additional hour)
slope= model.coef_[0]
print("slope (increase per additional hour):", slope)
```

```python
#Intercept(baseline monthly bill)                #model.intercept_ is the y-intercept, showing the starting value of the bill.
intercept= model.intercept_
print("Intercept (base bill):", intercept)
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting the data points 
plt.scatter(X, y, color= 'blue', label='Data Points')        #Scatter plot for actual data points

# Creating predictions for the entire range of X values for plotting the regression line
X_range = np.array([[x] for x in range(2, 14)])              #Generate X values for the range of interest
y_pred= model.predict(X_range)                       #Get predictions for the range of X values

#Plotting the regression line
plt.plot(X_range, y_pred, color= 'red', label= 'Regression Line')

#Highlight the predicted value for 5 hrs of daily usage
plt.scatter(5, pred_bill_5hrs, color= 'yellow', label='Prediction for 5 hours', s=100, edgecolor= 'black')

#Adding Labels and Titles
plt.title('Monthly Electricity Bill Prediction')
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill (dollars)')
plt.legend()
plt.grid(True)
plt.show()       #Show plot
```
