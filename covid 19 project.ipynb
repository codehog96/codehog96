```python
#finding out the missing values in each column of the dataset

missing_values= df.isnull().sum()
print(missing_values)
```

```python
#checking if there are any duplicate rows in the dataset

duplicates = df[df.duplicated()]
print(duplicates)
```

```python
#counting the total number of duplicated rows

duplicate_count = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_count}")
```

```python
#Detecting any outliers  #Using IQR(Interquartile Range) Method:
column1= 'Total Tests'
Q1= df[column1].quantile(0.25)
Q3= df[column1].quantile(0.75)
IQR = Q3-Q1
```

```python
lower_bound = Q1- 1.5*IQR
upper_bound = Q3+ 1.5*IQR
outliers = df[(df[column1] < lower_bound) | (df[column1] > upper_bound)]
print(outliers)
```

```python
#Replacing Outliers with Median:
median = df[column1].median()
df[column1] = np.where((df[column1] < lower_bound) | (df[column1] > upper_bound), median, df[column1])
df
```

```python
#Handling missing values
#By imputation method: using mean to fill in the missing values
median_Total_Recovered= df['Total Recovered'].median()
median_Total_Recovered
```

```python
#Fill missing Values in 'Total Recovered' with the mean of the column
df['Total Recovered'] = df['Total Recovered'].fillna(median_Total_Recovered)
df
```

```python
#filling the missing values in Active cases column with the median of the column
median_Active_Cases= df['Active Cases'].median()
median_Active_Cases
```

```python
df['Active Cases']= df['Active Cases'].fillna(median_Active_Cases)
df
```

```python
#Checking the cleaned dataset to make sure
df.info()
```

```python
#saving the cleaned dataframe to csv file
df.to_csv('cleaned_covid_dataset.csv', index=False)
```
